```python
#bins for age
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
players["Age Demographics"] = pd.cut(players["Age"], bins, labels=age_labels)
players
```

```python
Age_Demographics = players["Age Demographics"].value_counts()
Percentage_age = (Age_Demographics / Total_Players) * 100
Percentage_age 
```

# Age Demographics

```python
Age_Demographics_df = pd.DataFrame({"Total Count" : Age_Demographics,
                                   "Percentage of Players" : Percentage_age})
Age_Demographics_df = Age_Demographics_df.sort_index()
Age_Demographics_df
```

```python
pymoli_df["Purchase by Age"] = pd.cut(pymoli_df["Age"], bins, labels=age_labels)
pymoli_df
```

```python
total_purchase_price_by_age = pymoli_df.groupby(["Purchase by Age"]).sum()["Price"]
average_purchase_price_by_age = pymoli_df.groupby(["Purchase by Age"]).mean()["Price"]
purchase_count_by_age = pymoli_df.groupby(["Purchase by Age"]).count()["Price"]
average_total_by_person = total_purchase_price_by_age / Age_Demographics_df["Total Count"]
```

# Purchasing Analysis (Age)

```python
Purchasing_Analysis_Age = pd.DataFrame({"Purchase Count" : purchase_count_by_age, 
                                        "Average Purchase Price" : average_purchase_price_by_age,
                                       "Total Purchase Value" : total_purchase_price_by_age,
                                       "Avg Total Purchase per Person" : average_total_by_person})
Purchasing_Analysis_Age
#continue with line 104
```

```python
top_spenders_name = pymoli_df.groupby(["SN"]).count()["Purchase ID"]
average_purchase_spenders = pymoli_df.groupby(["SN"]).mean()["Price"]
total_purchase_spenders = pymoli_df.groupby(["SN"]).sum()["Price"]

Top_Spenders = pd.DataFrame({"Purchase Count" : top_spenders_name, 
                            "Average Purchase Price" : average_purchase_spenders,
                           "Total Purchase Value" : total_purchase_spenders})
Top_Spenders
#Age_Demographics_df.sort_index()
```

# Top Spenders

```python
Top_Spenders = Top_Spenders.sort_values("Total Purchase Value", ascending=False)
Top_Spenders
```

```python
#Most Popular Items
most_popular_items = pymoli_df.groupby(["Item ID", "Item Name"]).count()["Purchase ID"]
popular_items_price = pymoli_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_purchase_item = pymoli_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

Most_Popular_Items = pd.DataFrame({"Purchase Count" : most_popular_items,
                                  "Item Price" : popular_items_price,
                                  "Total Purchase Value" : total_purchase_item})
Most_Popular_Items
```

# Most Popular Items

```python
Most_Popular_Items = Most_Popular_Items.sort_values("Purchase Count", ascending=False)
Most_Popular_Items
```

# Most Profitable Items

```python
Most_Popular_Items = Most_Popular_Items.sort_values("Total Purchase Value", ascending=False)
Most_Popular_Items
```
